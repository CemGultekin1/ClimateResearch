```python
import xarray as xr
from scipy.ndimage import gaussian_filter
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import argparse
import intake
from intake import open_catalog
from matplotlib import colors, cm, pyplot as plt
```

Loading the dataset

```python
raw_data_address='/scratch/zanna/data/cm2.6/surf-data.zarr'
u_v_dataset=xr.open_zarr(raw_data_address).chunk(chunks={"time":1}).drop('surface_temp')
```

Computing "grid_data" that contains grid separation data

```python
x=u_v_dataset.xu_ocean.values
y=u_v_dataset.yu_ocean.values
dx=x[1:]-x[:-1]
dy=y[1:]-y[:-1]

dx=np.reshape(dx,(1,-1))
dx=[dx for i in range(len(y)-1)]
dx=np.concatenate(dx,axis=0)
#dx=np.stack([dx],axis=0)

dy=np.reshape(dy,(-1,1))
dy=[dy for i in range(len(x)-1)]
dy=np.concatenate(dy,axis=1)
grid_data=xr.Dataset(data_vars=dict(dxu=(["yu_ocean","xu_ocean"],dx),\
                                    dyu=(["yu_ocean","xu_ocean"],dy)),\
                        coords=dict(xu_ocean=x[:-1],yu_ocean=y[:-1]))
```

```python
grid_data
```

Taking derivatives

```python
dxu = grid_data['dxu']
dyu = grid_data['dyu']
gradient_x = u_v_dataset.diff(dim='xu_ocean') / dxu
gradient_y = u_v_dataset.diff(dim='yu_ocean') / dyu
```

```python
def get_grid(dataset,extremums):
    # outputs the grid data within certain values
    xmin,xmax,ymin,ymax=extremums
    dat=dataset.sel(xu_ocean=slice(xmin,xmax),yu_ocean=slice(ymin,ymax))
    x=dat.xu_ocean.values.reshape([-1,])
    y=dat.yu_ocean.values.reshape([-1,])

    xv, yv = np.meshgrid(x, y, indexing='ij')
    xv=xv.reshape([-1,])
    yv=yv.reshape([-1,])
    return xv,yv
```

```python
fig,axs=plt.subplots(1,1,figsize=(10,10))
fig.patch.set_facecolor('white')


xmin=-0.25
dx=2
xmax=xmin+dx
ymin=0
dy=2
ymax=ymin+dy
extremums=[xmin,xmax,ymin,ymax]

ax=axs

xv,yv=get_grid(gradient_x,extremums)
ax.scatter(xv,yv,color='b',label='gradient_x-grid',s=200,alpha=0.1)

xv,yv=get_grid(gradient_y,extremums)
ax.scatter(xv,yv,color='r',label='gradient_y-grid',s=60,alpha=0.1)

xv,yv=get_grid(u_v_dataset,extremums)
ax.scatter(xv,yv,color='k',label='original-grid',s=1)

ax.legend(loc='upper center')
ax.set_ylabel('latitude ($^{\circ}$)')
ax.set_xlabel('longitude ($^{\circ}$)')
```

```python
# Interpolate back the gradients
interp_coords = dict(xu_ocean=u_v_dataset.coords['xu_ocean'],
                     yu_ocean=u_v_dataset.coords['yu_ocean'])
gradient_x_interpolated = gradient_x.interp(interp_coords)
gradient_y_interpolated = gradient_y.interp(interp_coords)
```

```python
gradient_x_interpolated=gradient_x_interpolated.isel(time=0).sel(xu_ocean=slice(xmin,xmax),yu_ocean=slice(ymin,ymax))
gradient_y_interpolated=gradient_y_interpolated.isel(time=0).sel(xu_ocean=slice(xmin,xmax),yu_ocean=slice(ymin,ymax))
```
