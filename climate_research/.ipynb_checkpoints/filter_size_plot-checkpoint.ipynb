```python
def longitude_average(R2):
    mr2=np.zeros((R2.shape[0],R2.shape[1]))
    sr2=np.zeros((R2.shape[0],R2.shape[1]))
    for t in range(R2.shape[0]*R2.shape[1]):
        i=t%R2.shape[0]
        j=t//R2.shape[0]
        rr=R2[i,j]
        rr=rr[rr>0]
        rr[rr>1]=1
        with np.errstate(divide='ignore'):
            if len(rr)>0:
                mr2[i,j]=np.mean(rr)
                sr2[i,j]=np.std(rr)
    return mr2,sr2            
def r2fun(MSE,SC2):
    ZZ=SC2*1
    SC2[SC2==0]=1
    ZZ[ZZ!=0]=1
    ZZ[ZZ==0]=np.nan
    R2=1-MSE/SC2
    R2=R2*ZZ
    return longitude_average(R2)

def r2val(MSE,SC2):
    ZZ=SC2*1
    SC2[SC2==0]=1
    ZZ[ZZ!=0]=1
    ZZ[ZZ==0]=np.nan
    R2=1-MSE/SC2
    R2=R2*ZZ
    mr2=np.zeros(R2.shape[0])
    sr2=np.zeros(R2.shape[0])
    for t in range(R2.shape[0]):
        i=t
        rr=R2[i]
        rr=rr[rr>0]
        rr[rr>1]=1
        with np.errstate(divide='ignore'):
            if len(rr)>0:
                mr2[i]=np.mean(rr)
                sr2[i]=np.std(rr)
    return mr2,sr2        
```

```python
def indplacement(tvals,colsep):
    if type(colsep)==int:
        rowsep=[colsep]
    if len(colsep)==0:
        return [0 for i in range(len(tvals))]
    colset=[]
    idx=[]
    for j in range(len(colsep)):
        colcont=[]
        idx_=[]
        for i in range(len(tvals)):
            if not tvals[i][colsep[j]] in colcont:
                colcont.append(tvals[i][colsep[j]])
                idx_.append(len(colcont)-1)
            else:
                idx_.append(list(filter(lambda s: tvals[i][colsep[j]]==colcont[s],range(len(colcont))))[0])
        colset.append(colcont)
        idx.append(idx_)
    colnums=[1]
    for i in range(len(colset)):
        colnums.append(colnums[-1]*len(colset[i]))
    idx0=[]
    for i in range(len(idx[0])):
        colnum=0
        for j in range(len(idx)):
            colnum+=idx[j][i]*colnums[j]
        idx0.append(colnum)
    return idx0
```

```python
def try_get_r2(filename):
    try:
        MSE=np.load(filename)
        SC2=np.load(filename.replace('MSE','SC2'))
        R2,_=r2val(MSE,SC2)
        return R2
    except:
        print('doesnt exist: '+filename)
        return np.zeros(3)*np.nan
```

```python
def try_plot_r2(ax,filename,sigma,output_id,clr_id,lbl,grids):
    try:
        MSE=np.load(filename)
        SC2=np.load(filename.replace('MSE','SC2'))
        R2,_=r2fun(MSE,SC2)
        sp1=(len(grids[sigma]['lat'])-R2.shape[1])//2
        sp2=len(grids[sigma]['lat'])-R2.shape[1]-sp1
        latvals=grids[sigma]['lat'][sp1:]
        if sp2>0:
            latvals=latvals[:-sp2]
        ax.plot(latvals,R2[output_id],clrs[clr_id],label=lbl)
    except:
        print('doesnt exist: '+filename)
```

```python
def code2vals(num,C):
    T=[]
    for i in range(len(C)):
        T.append(num%C[i])
        num=num//C[i]
    return T
```
