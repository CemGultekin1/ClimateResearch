```python
def get_sep_fun(sigma):
    if sigma>0:
        dszarr=xr.open_zarr('/scratch/zanna/data/cm2.6/coarse-surf-data-sigma-'+str(sigma)+'.zarr')
    else:
        dszarr=xr.open_zarr('/scratch/zanna/data/cm2.6/surf-data.zarr')
    yu=dszarr.yu_ocean.values#.sel(yu_ocean=slice(-85,85))
    xu=dszarr.xu_ocean.values#.sel(yu_ocean=slice(-85,85))
    radius=6378.1
    dx=xu[1]-xu[0]
    dy=yu[1:]-yu[:-1]
    dx_=[]
    dy_=[]
    for i in range(len(yu)-1):
        y=yu[i]
        side_x=np.sin(np.pi*dx/180)*2*radius*np.sin(np.pi*(0.5-np.abs(y)/180))
        side_y=radius*np.sin(np.pi*(0.5-np.abs(y)/180))*dy[i]/180*np.pi
        dx_.append(side_x)
        dy_.append(side_y)
    dx_.append(dx_[-1])
    dy_.append(dy_[-1])
    return yu,dx_,dy_
```

```python

material=[]
rnum=10
    
for j in range(4):
    sigma=sigmas[j]
    yu,dx_,dy_=get_sep_fun(0)
    dy_=np.array(dy_)[np.arange(len(dy_)//sigma)*sigma]
    yu=np.array(yu)[np.arange(len(yu)//sigma)*sigma]
    Qx=Qys[j]
    
    u,s,vh=np.linalg.svd(Qx)
    IQx=u[:,:len(s)]@np.diag(1/s)@vh
    E=np.zeros((IQx.shape[0],rnum))
    mx=IQx.shape[1]
    print(sigma,s[-1]/s[0],IQx.shape)
    for i in range(IQx.shape[0]):
        iqcx=np.abs(IQx[i])
        ai=np.argmax(iqcx)
        I=np.mod(np.arange(mx)+ai,mx)-mx//2
        iqcx=iqcx[I]
        rh=iqcx[mx//2:]**2
        lh=iqcx[:mx//2+1]**2
        rh=np.sqrt(np.cumsum(rh[::-1]))[::-1]
        lh=np.sqrt(np.cumsum(lh))[::-1]
        if len(lh)>len(rh):
            rh=np.concatenate([rh,np.zeros(1)],axis=0)
        elif len(rh)>len(lh):
            lh=np.concatenate([lh,np.zeros(1)],axis=0)
        hh=np.maximum(lh,rh)
        '''iqcx[mx//2+1:]=iqcx[mx//2+1:]
        iqcx=(iqcx+iqcx[::-1])/2'''
        yuI=yu[I]
        yuI=np.abs(yuI[mx//2]-yuI)
        #E[i,0]=np.sum(iqcx**2)
        for r in range(E.shape[1]):
            E[i,r]=np.sum(iqcx[mx//2+r:]**2)+np.sum(iqcx[:mx//2-r]**2)
        IQx[i]=iqcx#/iqcx[mx//2]
    E=np.sort(E,axis=0)
    mx=E.shape[0]
    E=E[np.array([mx//4,mx//2,mx//4*3])]
    material.append(E)
```

```python
material=[]
cutoffs=[1e-1,1e-2,1e-3]

def findcutoffs(qq,cutval):
    i=0
    while qq[i]>cutval:
        i+=1
    return i
#     cut1=i
    
#     i=0
#     while qq[mm-i]>cutval:
#         i+=1
    return np.maximum(cut1,i)
for j in range(4):
    
    sigma=sigmas[j]
    
    yu,dx_,dy_=get_sep_fun(0)
    dy_=np.array(dy_)[np.arange(len(dy_)//sigma)*sigma]
    yu=np.array(yu)[np.arange(len(yu)//sigma)*sigma]
    Qx=Qys[j]
    
    u,s,vh=np.linalg.svd(Qx)
    IQx=u[:,:len(s)]@np.diag(1/s)@vh
    cuts=np.zeros((IQx.shape[0],len(cutoffs)))
    physlocs=np.zeros((IQx.shape[0],len(cutoffs)))
    D=IQx*0
    mx=IQx.shape[1]
    
    print(sigma,s[-1]/s[0],IQx.shape)
    for i in range(IQx.shape[0]):
        iqcx=np.abs(IQx[i])
        ai=np.argmax(iqcx)
        I=np.mod(np.arange(mx)+ai,mx)-mx//2
        iqcx=iqcx[I]
        rh=iqcx[mx//2:]**2
        lh=iqcx[:mx//2+1]**2
        rh=np.sqrt(np.cumsum(rh[::-1]))[::-1]
        lh=np.sqrt(np.cumsum(lh))[::-1]
        if len(lh)>len(rh):
            rh=np.concatenate([rh,np.zeros(1)],axis=0)
        elif len(rh)>len(lh):
            lh=np.concatenate([lh,np.zeros(1)],axis=0)
        hh=np.maximum(lh,rh)
        iqcx[m//2+1:]=iqcx[m//2+1:]
        iqcx=(iqcx+iqcx[::-1])/2
        yuI=yu[I]
        yuI=np.abs(yuI[mx//2]-yuI)
        IQx[i]=iqcx#/iqcx[mx//2]
        cuts[i]=np.array([findcutoffs(hh,cutoffs[k]) for k in range(len(cutoffs))]).astype(int)
        physlocs[i]=np.minimum(yuI[mx//2+cuts[i].astype(int)],yuI[mx//2-cuts[i].astype(int)])
    material.append([cuts,physlocs])
```
