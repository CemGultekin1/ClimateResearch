```python
import xarray as xr
import torch
import numpy as np
import os
import json
import copy
import numpy as np
import shutil
from scipy.io import netcdf


from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.lines import Line2D
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
from cartopy.mpl.geoaxes import GeoAxes
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from mpl_toolkits.axes_grid1 import AxesGrid



# os.system("jupyter nbconvert --to script 'climate_train.ipynb'")
# os.system("jupyter nbconvert --to script 'climate_data.ipynb'")
# os.system("jupyter nbconvert --to script 'climate_models.ipynb'")
# os.system("jupyter nbconvert --to script 'climate_job_maintenance.ipynb'")

import climate_train
import climate_data
import climate_models
import climate_job_maintenance as cjm

def longitude_average(R2):
    mr2=np.zeros((R2.shape[0],R2.shape[1]))
    sr2=np.zeros((R2.shape[0],R2.shape[1]))
    for t in range(R2.shape[0]*R2.shape[1]):
        i=t%R2.shape[0]
        j=t//R2.shape[0]
        rr=R2[i,j]
        rr=rr[rr>0]
        rr[rr>1]=1
        with np.errstate(divide='ignore'):
            if len(rr)>0:
                mr2[i,j]=np.mean(rr)
                sr2[i,j]=np.std(rr)
    return mr2,sr2
def longitude_full_average(R2):
    mr2=np.zeros(R2.shape[0])
    sr2=np.zeros(R2.shape[0])
    for t in range(R2.shape[0]):
        rr=R2[t]
        rr=rr[rr>0]
        rr[rr>1]=1
        with np.errstate(divide='ignore'):
            if len(rr)>0:
                mr2[t]=np.mean(rr)
                sr2[t]=np.std(rr)
    return mr2,sr2
def code2vals(num,C):
    T=[]
    for i in range(len(C)):
        T.append(num%C[i])
        num=num//C[i]
    return T
def indplacement(tvals,colsep):
    if type(colsep)==int:
        rowsep=[colsep]
    if len(colsep)==0:
        return [0 for i in range(len(tvals))]
    colset=[]
    idx=[]
    for j in range(len(colsep)):
        colcont=[]
        idx_=[]
        for i in range(len(tvals)):
            if not tvals[i][colsep[j]] in colcont:
                colcont.append(tvals[i][colsep[j]])
                idx_.append(len(colcont)-1)
            else:
                idx_.append(list(filter(lambda s: tvals[i][colsep[j]]==colcont[s],range(len(colcont))))[0])
        colset.append(colcont)
        idx.append(idx_)
    colnums=[1]
    for i in range(len(colset)):
        colnums.append(colnums[-1]*len(colset[i]))
    idx0=[]
    for i in range(len(idx[0])):
        colnum=0
        for j in range(len(idx)):
            colnum+=idx[j][i]*colnums[j]
        idx0.append(colnum)
    return idx0
def xyaxis_select(tvals,tstr,linetype,r2vals,rest,xaxis,idx):
    tvals_red=[[tvals[idx[s]][r] for r in rest] for s in range(len(idx))]
    tstr_red=[[tstr[idx[s]][r] for r in rest] for s in range(len(idx))]
    tvals_=[tvals[idx[s]] for s in range(len(idx))]
    linetype_=[linetype[idx[s]] for s in range(len(idx))]
    tstr_=[tstr[idx[s]] for s in range(len(idx))]
    r2vals_=[r2vals[idx[s]] for s in range(len(idx))]
    y=[]
    x=[]
    tval_col=[]
    lbl_col=[]
    line=[]
    for i in range(len(tvals_red)):
        if not tvals_red[i] in tval_col:
            tval_col.append(tvals_red[i])
            x.append([tstr_[i][xaxis]])
            y.append([r2vals_[i]])
            lbl=''
            for word in tstr_red[i]:
                lbl+=word+' '
            lbl_col.append(lbl)
            line.append(linetype_[i])
        else:
            line_index=list(filter(lambda s: tvals_red[i]==tval_col[s],range(len(tval_col))))[0]
            x[line_index].append(tstr_[i][xaxis])
            y[line_index].append(r2vals_[i])
    return x,y,lbl_col,line

def line_select(tvals,rest):
    tvals_red=[[tvals[i][r] for r in rest] for i in range(len(tvals))]
    tvals_=tvals.copy()
    tval_col=[]
    linetype=[]
    cur=0
    for i in range(len(tvals_red)):
        if not tvals_red[i] in tval_col:
            tval_col.append(tvals_red[i])
            linetype.append(cur)
            cur+=1
        else:
            li=list(filter(lambda s: tvals_red[i]==tval_col[s],range(len(tval_col))))[0]
            linetype.append(li)
    return linetype



def big_plot(modelnums,addpkp,C,\
             archtype=['LCNN','QCNN','UNET','PKP','REG'],d1=20,d2=20):
    root='/scratch/cg3306/climate/runs/'
    root1='/scratch/cg3306/climate/projection_analysis/'
    if addpkp:
        pkploc=0
        for i in range(len(archtype)):
            if archtype[i]=='PKP':
                pkploc=i
                break

    outputtype=['Subgrid Momentum Forcing (u)',\
               'Subgrid Momentum Forcing (v)',\
               'Subgrid Temperature Forcing']
    datatype=['Surface',\
               'Deep (110 m)']
    
    sigmavals=[4,8,12,16]
    sigmavals=[str(s) for s in sigmavals]
    geophys=['','glbl','glbl+coords']
    res=['res','']
    co2lvl=['','+1%CO2']


    runs=['G-'+str(i) for i in modelnums]
    modelnums=[num%1000 for num in modelnums]

    attribute_names=['data-type','arch-type','sigma','geophys','res','co2','output-type']
    added_attributes=[5,6]
    attributes_=[datatype,archtype,sigmavals,geophys,res,co2lvl,outputtype]
    attribute_ranks={}
    attributes={}
    for i in range(len(attribute_names)):
        attributes[attribute_names[i]]=attributes_[i]
        attribute_ranks[attribute_names[i]]=i



    tvals=[code2vals(num,C) for num in modelnums]
    tvals1=copy.deepcopy(tvals)
    for i in range(len(added_attributes)):
        natt=len(attributes_[added_attributes[i]])
        tvals2=[t+[0] for t in tvals1]
        for j in range(1,natt):
            tvals2+=[t+[j] for t in tvals1]
        tvals1=copy.deepcopy(tvals2)
    runs=runs*(len(tvals1)//len(tvals))
    tvals=tvals1

    rowsep=[attribute_ranks['data-type'],attribute_ranks['co2']]#[attribute_ranks['data-type'],attribute_ranks['res'],attribute_ranks['co2']]
    colsep=[attribute_ranks['output-type']]
    latitude_grids={}
    for str_sigma in sigmavals:
        sigma=int(str_sigma)
        ds_data=xr.open_zarr('/scratch/zanna/data/cm2.6/coarse-surf-data-sigma-'+str(sigma)+'.zarr')
        latitude_grids[str_sigma]=ds_data.yu_ocean.values
    num_file=len(tvals)
    source_file=[]
    label=[]
    latitude_idx=[]
    for i in range(num_file):
        tt=tvals[i]
        sigma=sigmavals[tt[attribute_ranks['sigma']]]
        data_id=tt[attribute_ranks['data-type']]
        co2_id=tt[attribute_ranks['co2']]
        label.append('sigma='+str(sigma)+' '+datatype[data_id]+' '+co2lvl[co2_id])
        latitude_idx.append(tt[attribute_ranks['sigma']])
        if co2_id==0:
            source_file.append(root+runs[i]+'/MSE.npy')
        else:
            source_file.append(root+runs[i]+'/MSE-co2.npy')
    irow=indplacement(tvals,rowsep)
    icol=indplacement(tvals,colsep)
    if addpkp:
        ncol=np.max(icol)+1
        nrow=np.max(irow)+1
        for i in range(nrow):
            for j in range(ncol):
                tt=tvals[np.where((np.array(irow)==i) * (np.array(icol)==j))[0][0]]
                filename=root1+''
                if tt[attribute_ranks['co2']]==1:
                    filename+='1pct-CO2-'
                if tt[attribute_ranks['data-type']]==0:
                    filename+='surf-data-'
                else:
                    filename+='3D-data-'
                filename+='sigma-'
                for k in range(len(sigmavals)):
                    filename1=filename+sigmavals[k]+'MSE.npy'
                    tt1=tt.copy()
                    tt1[attribute_ranks['arch-type']]=pkploc
                    tt1[attribute_ranks['sigma']]=k
                    tt1[attribute_ranks['res']]=1
                    tvals.append(tt1)
                    source_file.append(filename1)
                    irow.append(i)
                    icol.append(j)
    tstr=[[attributes_[i][t[i]] for i in range(len(t)) ] for t in tvals]
    latitude_plot=False
    r2vals=[]
    for k in range(len(source_file)):
        i=irow[k]
        try:
            with np.errstate(divide='ignore'):
                MSE=np.load(source_file[k])
                SC2=np.load(source_file[k].replace('MSE','SC2'))
                ZZ=SC2*1
                SC2[SC2==0]=1
                ZZ[ZZ!=0]=1
                ZZ[ZZ==0]=np.nan
                R2=1-MSE/SC2
                R2=R2*ZZ
            mr2,sr2=longitude_full_average(R2)
            r2vals.append(mr2[tvals[k][attribute_ranks['output-type']]])
        except:
            searched_file=source_file[k]
            pth_exists=os.path.exists(searched_file)
            print(str(k)+': '+searched_file+' exists: '+str(pth_exists))
            r2vals.append([np.nan])
    rest=list(range(len(attribute_names)))
    xaxis=attribute_ranks['sigma']
    core=[i for i in rest if i in colsep or i in rowsep and i!=xaxis]
    rest=[i for i in rest if not i in colsep and not i in rowsep and i!=xaxis]
    ncol=np.max(icol)+1
    nrow=np.max(irow)+1
    linetype=line_select(tvals,rest)



    clrs=['blue','red','coral','green','purple','magenta','darkblue','teal']
    mrkrs=Line2D.filled_markers


    fig,axs=plt.subplots(nrow,ncol,figsize=(d2,d1))
    fig.patch.set_facecolor('white')
    for i in range(nrow):
        for j in range(ncol):
            ax=axs[i,j]
            idx=[ir for ir in range(len(irow)) if irow[ir]==i and icol[ir]==j and not np.any(np.isnan(r2vals[ir]))]

            x,y,lbl,line=xyaxis_select(tvals,tstr,linetype,r2vals,rest,xaxis,idx)

            for linenum in range(len(x)):
                x_=x[linenum]
                y_=y[linenum]
                x_=np.array([int(x_[i]) for i in range(len(x_))])
                y_=np.array(y_)
                I=np.argsort(x_)
                x_=x_[I]
                y_=y_[I]
                ax.plot(x_,y_,color=clrs[line[linenum]],label=lbl[linenum])#
                ax.plot(x_,y_,'k'+mrkrs[0])
            title=''
            for kk in core:
                title+=tstr[idx[0]][kk]+' '
            ax.legend()
            ax.grid(color='k',linestyle='--',alpha=0.5)
            #ax.set_ylim([0,1])
            ax.set_title(title)
            ax.set_xlabel('sigma')
            ax.set_ylabel('R2')
```

```python
def cartopy_subplot(fig,nrow,ncol,index,lons,lats,u,title,exts=[]):
    ax = plt.subplot(nrow,ncol,index, projection=ccrs.Robinson())
    if len(exts)==0:
        umax=u[u==u].max()
        umin=u[u==u].min()
    else:
        umin,umax=exts[0],exts[1]
    ds=xr.Dataset(data_vars=dict(u=(["latitude","longitude"],u)),\
                            coords=dict(latitude=lats,longitude=lons))

    cax=ds.u.plot(ax=ax, transform = ccrs.PlateCarree(),\
                  cmap='coolwarm',\
                  norm=mcolors.LogNorm(vmin=umin,vmax=umax),\
                  add_colorbar=False)
    #ax.set_xticks([-10,-1,0.1,0,0.1,1,10])#([-180,-90,0,90,180])
    #ax.set_yticks([-90,-60,-30,0,30,60,90])
    ax.set_title(title)
    ax.coastlines()
    
    tickvals=[10**(-(i-1)) for i in range(8)]
    tickstr=[str(tt) for tt in tickvals]
    cbar = fig.colorbar(cax, ticks=tickvals,fraction=0.035, pad=0.04)
    cbar.ax.set_yticklabels(tickstr)
    cbar.ax.set_ylabel('1e-7 m/s2')
```
